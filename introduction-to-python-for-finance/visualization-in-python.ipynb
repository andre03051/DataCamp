```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('data/stock_data (1).csv')
df.head()
```

```python
df2 = pd.read_csv('data/stock_data2.csv')
df2.head()
```

```python
days = df.Day
prices = df.Price

# Plot the price of stock over time
plt.plot(days, prices, color="red", linestyle="--")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot price as a function of time
plt.plot(days, prices, color="red", linestyle="--")

# Add x and y labels
plt.xlabel('Days')
plt.ylabel('Prices, $')

# Add plot title
plt.title('Company Stock Prices Over Time')

# Show plot
plt.show()
```

```python
prices1 = prices[0:len(df2.company2)]
prices2 = df2.company2
days = df2.day

# Plot two lines of varying colors 
plt.plot(days, prices1, color='red')
plt.plot(days, prices2, color='green')

# Add labels
plt.xlabel('Days')
plt.ylabel('Prices, $')
plt.title('Stock Prices Over Time')
plt.show()
```

```python
days = df.Day
prices = df.Price

# Plot price as a function of time
plt.scatter(days, prices, color='green', s=0.1)

# Show plot
plt.show()
```

```python
# Plot histogram 
plt.hist(prices, bins=100)

# Display plot
plt.show()
```

```python
# Plot histogram of stocks_A
plt.hist(df2.company1, bins=100, alpha=.4)

# Plot histogram of stocks_B 
plt.hist(df2.company2, bins=100, alpha=.4)

# Display plot
plt.show()
```

```python
# Plot stock_A and stock_B histograms
plt.hist(df2.company1, bins=100, alpha=0.4, label='Stock A')
plt.hist(df2.company2, bins=100, alpha=0.4, label='Stock B')

# Add the legend
plt.legend()

# Display the plot
plt.show()
```
