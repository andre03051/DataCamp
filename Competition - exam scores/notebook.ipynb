```python
ml_data = df.copy()

ml_data['genderEnum'] = 0
ml_data.loc[ml_data['gender'] == 'female', 'genderEnum'] = 1
ml_data.drop('gender', axis=1, inplace=True)

ml_data['pelEnum'] = 0
ml_data.loc[ml_data['parent_education_level'] == 'high school', 'pelEnum'] = 1
ml_data.loc[ml_data['parent_education_level'] == 'some college', 'pelEnum'] = 2
ml_data.loc[ml_data['parent_education_level'] == "associate's degree", 'pelEnum'] = 3
ml_data.loc[ml_data['parent_education_level'] == "bachelor's degree", 'pelEnum'] = 4
ml_data.loc[ml_data['parent_education_level'] == "master's degree", 'pelEnum'] = 5
ml_data.drop('parent_education_level', axis=1, inplace=True)

ml_data['lunchEnum'] = 0
ml_data.loc[ml_data['lunch'] == 'standard', 'lunchEnum'] = 1
ml_data.drop('lunch', axis=1, inplace=True)

ml_data['test_prep_courseEnum'] = 0
ml_data.loc[ml_data['test_prep_course'] == 'completed', 'test_prep_courseEnum'] = 1
ml_data.drop('test_prep_course', axis=1, inplace=True)

ml_data = pd.get_dummies(ml_data, )

ml_data.head()
```

```python
X = ml_data.drop(['math', 'reading', 'writing'], axis=1)
y = ml_data['math']

X.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})

test_sizes = np.arange(.05, .5, .05)
train_score = []; set_score = []
for test_size in test_sizes:
    # Create the training and test sets
    X_train, X_test, y_train, y_test= train_test_split(
        X, 
        y, 
        test_size=test_size, 
        random_state = 123
    )

    # Instantiate dt
    y_pred_test = DecisionTreeClassifier(
        random_state=123
    ).fit(X_train, y_train).predict(X_test)

    # Evaluate accuracy
    train_score.append(
        accuracy_score(y_pred_test, y_test)
    )

figure, axes = plt.subplots(figsize = (18,6))
plt.plot(test_sizes, train_score)
plt.title('Accuracy Impact on Sample Size')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create the training and test sets
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=.35, random_state=123)

# Instantiate dt
rf = RandomForestClassifier(random_state=123)

# Fit bc to the training set
rf.fit(X_train, y_train)

# Predict test set labels
y_pred = rf.predict(X_test)

# Evaluate acc_test
acc_test = accuracy_score(y_pred, y_test)
print('Test set accuracy of RandomForest: {:.4f}'.format(acc_test)) 
```

```python
# Import AdaBoostClassifier
from sklearn.ensemble import AdaBoostClassifier

# Instantiate ada
ada = AdaBoostClassifier(
    base_estimator=rf,
    n_estimators=5,
    random_state=123
).fit(X_train, y_train)

# Predict test set labels
y_pred = ada.predict(X_test)

# Evaluate acc_test
acc_test = accuracy_score(y_pred, y_test)
print('Test set accuracy of ada: {:.4f}'.format(acc_test)) 
```

```python
from sklearn.ensemble import BaggingClassifier

# Instantiate bc
bc = BaggingClassifier(
    base_estimator=ada, 
    random_state=123,
    verbose=1,
    n_jobs = -1
)

# Fit bc to the training set
bc.fit(X_train, y_train)

# Predict test set labels
y_pred = bc.predict(X_test)

# Evaluate acc_test
acc_test = accuracy_score(y_pred, y_test)
print('Test set accuracy of BaggingClassifier: {:.4f}'.format(acc_test)) 
```
